1) Hallar la porción d de una esfera de radio $r=10cm$ y densidad $\rho=0,638 \frac{g}/{cm^3}​ $​ que queda sumergida en agua de manera estable.<br>
El modelo matemático es
$\frac{\pi(d^3​ ​-3d^2​r+4r^3​\rho )}{3} = 0$


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def f(d,r,rho):
    return (np.pi*(d**3 - 3*r*(d**2) + 4*(r**3) * rho))/3

a = -2
b = 3
d = np.arange(a,b,0.05)
r = 0.10
rho = 638
plt.plot(d,f(d,r,rho))
plt.grid()
plt.show()
```

```python
#la raiz es masomenos 0.3, entonces voy a buscar la raiz dentro del intervalo -0.1, 0.4

error= 10
i=0
while error >1e-7:
  c= (a+b)/2
  fa= f(a,r,rho)
  fc= f(c,r,rho)
  if fc==0:
    raiz = c
  elif fa*fc <0 :
    b = c
  else:
    a=c
  raiz=c
  i+=1
  error = abs(fc)

print(f'la raiz, es decir, la porcion de la pelota es: {round(raiz,5)}')
print(f'salió en  {i} iteraciones')
print(f'verificación: f({raiz}) = {round(f(raiz),3)}')
```

2) Cinemática elegante: Un tiro vertical a baja velocidad considerando la viscosidad del aire puede modelarse con la ecuación horaria (considerando $C=\frac{m}{\beta}$ siendo $\beta$ el coeficiente que multiplica la velocidad en la ecuación $F_v=\beta v$ para la fuerza de arrastre). Siendo $C=2s$, la altura inicial $h=10m$ y la velocidad de lanzamiento $18 \frac{m}{s}$, determinar el tiempo de vuelo.<br>
$h = h_0 + (Cv_0 + C^{2} g)(1 − e^{\frac{−t}{C}}) − gCt$

```python
import math 

def f2(t):
  C= 2 
  h=10
  v0=18
  g=9.8
  return h +(C * v0 + C ** 2 * g)* (1 - math.e ** (-t/C)) - g * C * t

t= np.arange(0,5,0.1)
plt.plot(t,f2(t))
plt.grid()
plt.show()

a = 0
b = 5
error= 10
i=0
while error >1e-8:
  c= (a+b)/2
  fa= f2(a)
  fc= f2(c)
  if fc==0:
    raiz = c
  elif fa*fc <0 :
    b = c
  else:
    a=c
  raiz=c
  i+=1
  error = abs(fc)

print(f'la raiz, es decir, el tiempo de vuelo es: {round(raiz,5)} segundos')
print(f'salió en  {i} iteraciones')
print(f'verificación: f({raiz}) = {round(f2(raiz),3)}')
```
