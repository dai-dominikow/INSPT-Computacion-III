### Machete para el parcial de C3: Como Plotear en 3D

```python
import matplotlib.pyplot as plt
import numpy as np 

# por convensin, se suele llamar a los grficos 3d como 'ax'
# para crear nuestro grafico y plotear sobre el distintos puntos, debemos crear un 'subplot' con proyeccion en 3 dimensiones

ax = plt.figure().add_subplot( projection='3d')

# para luego, como haciamos en 2d,  plotear un scatterplot con los parametros 'x','y' y 'z': 
x = np.array([0,-3,2,-1,2,3,-2])
y = np.array([1,0,2,1,-2,-3,1])
z = np.array([-1,2,-2,-1,-2,-2,-2])
ax.scatter(x,y,z) 
plt.show()
```

```python
#para plotear vectores, los parametros son los puntos de la ubicacion, y el campo vectorial en si
#por ejemplo, supongamos un campo M = (-y/4,2x,2z)
u = -y/4
v = 2*x
w = 2*z
ax = plt.figure().add_subplot( projection='3d')
ax.quiver(x,y,z,u,v,w)
plt.show()
```
